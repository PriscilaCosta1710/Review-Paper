```python
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import files

plt.rcParams['font.family'] = 'DejaVu Serif'

escalas = ['Regional', 'Global', 'Nacional', 'Local', 'Continental']
estudos = [30, 28, 25, 5, 4]

df = pd.DataFrame({'Escala': escalas, 'Estudos': estudos})

fig, ax = plt.subplots(figsize=(8, 5))
ax.bar(df['Escala'], df['Estudos'], color='cornflowerblue')

ax.set_xlabel('Escala de abrangência da área de estudo', fontsize=12, fontweight='bold')
ax.set_ylabel('Número de estudos', fontsize=12, fontweight='bold')
plt.title('Distribuição dos Estudos por Escala de Abrangência', fontsize=14, fontweight='bold')
ax.set_ylim(0, 35)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()

arquivo = 'distribuicao_por_escala.png'
plt.savefig(arquivo, dpi=300, bbox_inches='tight')
plt.show()
files.download(arquivo)
```
