```python
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import files

plt.rcParams['font.family'] = 'DejaVu Serif'

paises = ['Índia', 'Espanha', 'China', 'Brasil', 'EUA']
percentuais = [3.3, 3.3, 20.3, 26.4, 36.7]

df = pd.DataFrame({'País': paises, 'Percentual': percentuais})
df = df.sort_values(by='Percentual', ascending=True)

fig, ax = plt.subplots(figsize=(8, 5))
bars = ax.barh(df['País'], df['Percentual'], color='cornflowerblue')

ax.set_xlabel('Percentual de publicação (%)', fontsize=12, fontweight='bold')
ax.set_ylabel('País de publicação', fontsize=12, fontweight='bold')
plt.title('Distribuição de Publicações por País', fontsize=14, fontweight='bold')
ax.set_xlim(0, 40)
ax.set_xticks([0, 10, 20, 30, 40])  # Sem % no eixo

# Rótulo nas barras
for bar in bars:
    width = bar.get_width()
    ax.text(width + 1, bar.get_y() + bar.get_height()/2,
            f'{width:.1f}%', va='center', fontsize=10)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()

arquivo = 'publicacoes_por_pais.png'
plt.savefig(arquivo, dpi=300, bbox_inches='tight')
plt.show()
files.download(arquivo)
```
