```python
# Importar bibliotecas
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import files

# Configurar fonte semelhante a Times New Roman
plt.rcParams['font.family'] = 'DejaVu Serif'

# Dados reais (sem 2015)
years = [2012, 2013, 2014, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]
publications_per_year = [1, 1, 2, 2, 2, 6, 6, 4, 12, 15, 23, 18]
years_str = list(map(str, years))  # converter para string para não deixar espaço no eixo x

# Criar DataFrame
df = pd.DataFrame({
    'Year': years_str,
    'Publications_per_year': publications_per_year
})
df['Cumulative_total'] = df['Publications_per_year'].cumsum()

# Criar figura
fig, ax1 = plt.subplots(figsize=(10, 6))

# Barras por ano
ax1.bar(df['Year'], df['Publications_per_year'], color='cornflowerblue', label='Publications per year')
ax1.set_xlabel('Year', fontsize=12, fontweight='bold')
ax1.set_ylabel('Number of articles per year', fontsize=12)
ax1.set_ylim(0, 25)
ax1.tick_params(axis='y')

# Linha acumulada (segundo eixo)
ax2 = ax1.twinx()
ax2.plot(df['Year'], df['Cumulative_total'], color='darkorange', linewidth=2.5, label='Cumulative total')
ax2.set_ylabel('Cumulative number of articles', fontsize=12)
ax2.set_ylim(0, 100)
ax2.tick_params(axis='y')

# Estética
fig.tight_layout()
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')
plt.title('Evolution of Scientific Publications by Year and Cumulative Total', fontsize=14, fontweight='bold')
plt.grid(False)

# Salvar imagem em alta resolução (300 DPI)
plt.savefig('publications_graph.png', dpi=300, bbox_inches='tight')
plt.show()

# Baixar o arquivo diretamente no Colab
files.download('publications_graph.png')

```
