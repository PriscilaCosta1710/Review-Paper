```python
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import files

plt.rcParams['font.family'] = 'DejaVu Serif'

scales = ['Regional', 'Global', 'National', 'Local', 'Continental']
studies = [30, 28, 25, 5, 4]

df = pd.DataFrame({'Scale': scales, 'Studies': studies})

fig, ax = plt.subplots(figsize=(8, 5))
ax.bar(df['Scale'], df['Studies'], color='cornflowerblue')

ax.set_xlabel('Study area scale', fontsize=12, fontweight='bold')
ax.set_ylabel('Number of studies', fontsize=12, fontweight='bold')
plt.title('Distribution of Studies by Spatial Scale', fontsize=14, fontweight='bold')
ax.set_ylim(0, 35)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()

filename = 'scale_distribution.png'
plt.savefig(filename, dpi=300, bbox_inches='tight')
plt.show()
files.download(filename)
```
