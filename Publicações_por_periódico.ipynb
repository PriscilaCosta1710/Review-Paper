```python
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import files  # Somente se estiver usando Colab

# Configurar fonte estilo acadêmico
plt.rcParams['font.family'] = 'DejaVu Serif'

# Dados
periodicos = [
    'Journal of Hydrometeorology',
    'Atmosphere',
    'Environmental Research Letters',
    'Geophysical Research Letters',
    'Science of The Total Environment',
    'Bulletin of the American Meteorological Society',
    'Environmental Research Letters ',
    'Hydrology and Earth System Sciences',
    'International Journal of Climatology',
    'Remote Sensing',
    'Water Resources Research'
]

publicacoes = [9, 4, 4, 4, 4, 3, 3, 3, 3, 3, 3]

# Criar DataFrame e ordenar
df = pd.DataFrame({
    'Periódico': periodicos,
    'Publicações': publicacoes
})
df = df.sort_values(by='Publicações', ascending=True)

# Gráfico de barras horizontais
fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.barh(df['Periódico'], df['Publicações'], color='cornflowerblue')

# Estética e rótulos
ax.set_xlabel('Número de publicações', fontsize=12, fontweight='bold')
ax.set_ylabel('Nome do periódico', fontsize=12, fontweight='bold')
plt.title('Distribuição de Publicações por Periódico', fontsize=14, fontweight='bold')
plt.xlim(0, 10)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()

# Salvar e baixar automaticamente
nome_arquivo = 'publicacoes_periodicos.png'
plt.savefig(nome_arquivo, dpi=300, bbox_inches='tight')
plt.show()

# ⬇️ Download automático (funciona no Google Colab)
files.download(nome_arquivo)

```
