```python
# Importar bibliotecas
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import files

# Configurar fonte parecida com Times New Roman
plt.rcParams['font.family'] = 'DejaVu Serif'

# Dados reais (sem o ano de 2015)
anos = [2012, 2013, 2014, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]
publicacoes_por_ano = [1, 1, 2, 2, 2, 6, 6, 4, 12, 15, 23, 18]
anos_str = list(map(str, anos))  # Converter para string para evitar espaos no eixo x

# Criar DataFrame
df = pd.DataFrame({
    'Ano': anos_str,
    'Publicacoes_por_ano': publicacoes_por_ano
})
df['Total_acumulado'] = df['Publicacoes_por_ano'].cumsum()

# Criar figura
fig, ax1 = plt.subplots(figsize=(10, 6))

# Barras por ano
ax1.bar(df['Ano'], df['Publicacoes_por_ano'], color='cornflowerblue', label='Publicaes por ano')
ax1.set_xlabel('Ano', fontsize=12, fontweight='bold')
ax1.set_ylabel('Nmero de artigos por ano', fontsize=12)
ax1.set_ylim(0, 25)
ax1.tick_params(axis='y')

# Linha acumulada (eixo secundrio)
ax2 = ax1.twinx()
ax2.plot(df['Ano'], df['Total_acumulado'], color='darkorange', linewidth=2.5, label='Total acumulado')
ax2.set_ylabel('Nmero acumulado de artigos', fontsize=12)
ax2.set_ylim(0, 100)
ax2.tick_params(axis='y')

# Esttica
fig.tight_layout()
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')
plt.title('Evoluo das Publicaes Cientficas por Ano e Total Acumulado', fontsize=14, fontweight='bold')
plt.grid(False)

# Salvar imagem com alta resoluo (300 DPI)
plt.savefig('grafico_publicacoes.png', dpi=300, bbox_inches='tight')
plt.show()

# Baixar automaticamente o arquivo no Colab
files.download('grafico_publicacoes.png')

```
