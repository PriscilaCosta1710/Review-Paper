```python
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import files  # Apenas no Google Colab

# Configurar fonte acadêmica
plt.rcParams['font.family'] = 'DejaVu Serif'

# Dados dos autores
autores = [
    'Lowman, L.E.L.', 'Marengo, J.A.', 'Hoffmann, D.', 'Yuan, X.', 'Hobbins, M.',
    'Anderson, M.C.', 'Hain, C.R.', 'Hoell, A.', 'Mishra, V.', 'Ford, T.W.',
    'Barbosa, H.A.', 'Svoboda, M.', 'Christian, J.I.', 'Basara, J.B.', 'Hunt, E.D.', 'Otkin, J.A.'
]

publicacoes = [4, 4, 4, 5, 5, 5, 5, 5, 6, 6, 6, 7, 7, 8, 9, 15]

# Criar e ordenar DataFrame
df = pd.DataFrame({'Autor': autores, 'Publicações': publicacoes})
df = df.sort_values(by='Publicações', ascending=True)

# Gráfico de barras horizontais
fig, ax = plt.subplots(figsize=(10, 6))
ax.barh(df['Autor'], df['Publicações'], color='cornflowerblue')

# Rótulos e estética
ax.set_xlabel('Número de publicações', fontsize=12, fontweight='bold')
ax.set_ylabel('Nome do autor', fontsize=12, fontweight='bold')
plt.title('Distribuição de Publicações por Autor', fontsize=14, fontweight='bold')
plt.xlim(0, 16)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()

# Salvar e baixar
arquivo = 'publicacoes_autores.png'
plt.savefig(arquivo, dpi=300, bbox_inches='tight')
plt.show()
files.download(arquivo)
```
