```python
import matplotlib.pyplot as plt
import pandas as pd

# Lista real de anos sem 2015
anos = [2012, 2013, 2014, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]
publicacoes_por_ano = [1, 1, 2, 2, 2, 6, 6, 4, 12, 15, 23, 18]

# Converter ano para string para evitar interpolao contnua
anos_str = list(map(str, anos))

# Criar DataFrame
df = pd.DataFrame({'Ano': anos_str, 'Publicacoes_por_ano': publicacoes_por_ano})
df['Total_acumulado'] = df['Publicacoes_por_ano'].cumsum()

# Grfico
fig, ax1 = plt.subplots(figsize=(10, 6))
ax1.bar(df['Ano'], df['Publicacoes_por_ano'], color='cornflowerblue', label='Nmero de publicaes por ano')
ax1.set_xlabel('Ano', fontsize=12, fontweight='bold')
ax1.set_ylabel('Valores absolutos de artigos', fontsize=12)
ax1.set_ylim(0, 25)
ax1.tick_params(axis='y')

# Linha acumulada SEM crculos
ax2 = ax1.twinx()
ax2.plot(df['Ano'], df['Total_acumulado'], color='darkorange', linewidth=2.5, label='Total acumulado')
ax2.set_ylabel('Total acumulado de artigos', fontsize=12)
ax2.set_ylim(0, 100)
ax2.tick_params(axis='y')

# Esttica
fig.tight_layout()
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')
plt.title('Evoluo das Publicaes por Ano e Total Acumulado', fontsize=14, fontweight='bold')
plt.grid(False)
plt.show()
```
