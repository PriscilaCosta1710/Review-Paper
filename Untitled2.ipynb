```python
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import files  # Only for Colab

# Set academic font
plt.rcParams['font.family'] = 'DejaVu Serif'

# Data
countries = ['India', 'Spain', 'China', 'Brazil', 'USA']
percentages = [3.3, 3.3, 20.3, 26.4, 36.7]

# Create and sort DataFrame
df = pd.DataFrame({'Country': countries, 'Percentage': percentages})
df = df.sort_values(by='Percentage', ascending=True)

# Plot
fig, ax = plt.subplots(figsize=(8, 5))
bars = ax.barh(df['Country'], df['Percentage'], color='cornflowerblue')

# Axis labels and title
ax.set_xlabel('Publication percentage (%)', fontsize=12, fontweight='bold')
ax.set_ylabel('Country', fontsize=12, fontweight='bold')
plt.title('Distribution of Publications by Country', fontsize=14, fontweight='bold')
ax.set_xlim(0, 40)
ax.set_xticks([0, 10, 20, 30, 40])  # No percent symbol

# Add percentage labels next to bars
for bar in bars:
    width = bar.get_width()
    ax.text(width + 1, bar.get_y() + bar.get_height() / 2,
            f'{width:.1f}%', va='center', fontsize=10)

# Clean style
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()

# Save and download
filename = 'publications_by_country.png'
plt.savefig(filename, dpi=300, bbox_inches='tight')
plt.show()
files.download(filename)
```
